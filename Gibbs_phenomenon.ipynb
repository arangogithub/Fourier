```python
%matplotlib notebook
import numpy as np
from numpy import fft
import matplotlib.pyplot as plt
```

```python
def dft(n, x):
    odds = 2*np.arange(float(n)) + 1
    return 1/2.0+ np.sum((2/np.pi)*(1/odds)*np.sin(x*odds))

approx = np.vectorize(dft)
```

```python
xx = np.linspace(-np.pi, np.pi, 1000)
yy5 = approx(5,xx)
yy10 = approx(10,xx)
yy20 = approx(20,xx)

xxr =  np.linspace(0, np.pi, 100)
xxl =  np.linspace(-np.pi, 0, 100)
uno = np.ones(len(xxr))
cero = np.zeros(len(xxr))
```

```python
lista = [ np.amax(approx(n,xx)) for n in range(50)]
```

```python
plt.plot(lista)
plt.show()
```

```python
xlabels = [r'$-\pi$',r'$0$', r'$\pi$' ]
yylabels = [r'$-0.2$',r'$0.5$', r'$0.5$' ]
f, ax = plt.subplots(3,1,sharex=True)

ax[0].plot(xx, yy5)  

ax[0].plot(xxr, uno)
ax[0].plot(xxl, cero)
ax[0].set_yticks([-0.2, 0.5, 1.2])
ax[0].set_ylabel(r'$N = 5$', size = 'x-large')

#######################################
ax[1].plot(xx, yy10) 
ax[1].plot(xxr, uno)
ax[1].plot(xxl, cero)
ax[1].set_yticks([-0.2, 0.5, 1.2])
ax[1].set_ylabel(r'$N = 10$', size = 'x-large')
###############################################
ax[2].plot(xx, yy20) 
ax[2].plot(xxr, uno)
ax[2].plot(xxl, cero)
ax[2].set_yticks([-0.2, 0.5, 1.2])
ax[2].set_ylabel(r'$  N = 20  $', size = 'x-large')
plt.xticks([-np.pi, 0, np.pi], xlabels, size = 'x-large')
ax[2].set_xlabel(r'$x$', size = 'x-large')
#plt.show()
```
