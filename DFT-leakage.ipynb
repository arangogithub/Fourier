```python
%matplotlib notebook
import numpy as np
from numpy import fft
import matplotlib.pyplot as plt
```

```python
N = 2**12
dx = 0.1
x = dx*np.arange(0,N)
w1, w2 = 2**5, 2**3
N, w1, w2
fx = (dx*x+1)%2 -1
realdft = fft.rfft(fx)/N
nu = fft.rfftfreq(N,dx) # Natural frequencies
```

```python
realdft = fft.rfft(fx)/N
nu = fft.rfftfreq(N,dx) # Natural frequencies
```

```python
f, ax = plt.subplots(3,1,sharex=True)
ax[0].plot(nu, 2*np.real(realdft)) # Plot Cosine terms
ax[0].set_ylabel(r'$a_k$', size = 'x-large')
ax[1].plot(nu, -2*np.imag(realdft)) # Plot Sine terms
ax[1].set_ylabel(r'$b_k$', size = 'x-large')
ax[2].plot(nu, 4*np.absolute(realdft)**2) # Plot spectral power
ax[2].set_ylabel(r'$  a_k^2 + b_k^2  $', size = 'x-large')
ax[2].set_xlabel(r'$\widetilde{\nu}$', size = 'x-large')
plt.show()
```
